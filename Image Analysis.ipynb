```python
import numpy as np
import imageio
from matplotlib import pyplot as plt
rockimage = imageio.imread('berea8bit.tif')
plt.imshow(rockimage)
```

```python
plt.hist(np.reshape(rockimage, (500**2, 1)), density=True, bins=30, range=[0,255])
plt.title('Grain and Porosity Distribution')
plt.ylabel('n (normalized)') #Frequency
plt.xlabel ('Porosity and Grain Distribution')
# When plotting the left hand side is porosity and the right hand side is grain
plt.show()
```

```python
BW = np.zeros((500,500))
print(BW)
```

```python
bw = np.zeros((500,500))
print(bw)
```

```python
bw[rockimage > 80]= 1
bw2 = bw.astype(np.bool)
print(bw2)
```

```python
frame = np.sum(bw2)
print(frame)
```

```python
totalV= np.size(bw2)
print(totalV)
```

```python
pores = totalV - frame
print(pores)
```

```python
phi = ((totalV-frame)/ totalV)*100
print(phi)
```
